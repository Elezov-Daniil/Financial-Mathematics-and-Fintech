```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
x = 40
r = 0.1
sig = 0.5
T = 1
t = 0
h = np.linspace(0, 100, 101)
K = 20
```

```python
def get_d_plus(x, sig, T, t, r, K):
    d_plus = (np.log(x / K) + (r + sig ** 2 / 2) * (T - t)) / (sig * np.sqrt(T - t))
    return d_plus

def get_d_minus(x, sig, T, t, r, K):
    d_minus = (np.log(x / K) + (r - sig ** 2 / 2) * (T - t)) / (sig * np.sqrt(T - t))
    return d_minus

def price_option(x, sig, T, t, r, K):
    C_test = []
    for x in h:
        C1_test = x * stats.norm.cdf(get_d_plus(x, sig, T, t, r, K)) - K * np.exp( - r * (T - t)) * stats.norm.cdf(get_d_minus(x, sig, T, t, r, K))
        C_test.append(C1_test)
    return C_test
```

```python
plt.plot(h, price_option(x, sig, T, 0.3, r, K), label='t = 0')
plt.plot(h, price_option(x, sig, T, 0.5 * T, r, K), label='t = 0.5')
plt.plot(h, price_option(x, sig, T, 0.7 * T, r, K), label='t = 0.8')
plt.plot(h, price_option(x, sig, T, 0.9 * T, r, K), label='t = 0.9')
plt.legend()
plt.grid()
```

```python
def delta(x, sig, T, t, r, K):
    delta = []
    for x in h: 
        delta.append(stats.norm.cdf(get_d_plus(x, sig, T, t, r, K)))
    return delta

def gamma(x, sig, T, t, r, K):
    gamma = []
    for x in h: 
        gamma1 = np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) / (x * sig * np.sqrt(T - t) * np.sqrt(2 * np.pi))
        gamma.append(gamma1)
    return gamma

def theta(x, sig, T, t, r, K):
    theta = []
    for x in h: 
        theta1 = - x * sig * np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) * np.sqrt(T - t) / (np.sqrt(2 * np.pi) * 2) - r * K * np.exp( - r * (T - t) * stats.norm.cdf(get_d_minus(x, sig, T, t, r, K)))
        theta.append(theta1)
    return theta

def vega(x, sig, T, t, r, K):
    vega = []
    for x in h: 
        vega1 = (x * np.sqrt(T - t) * np.exp( -1 * get_d_plus(x, sig, T, t, r, K) ** 2 / 2))  / np.sqrt(2 * np.pi)
        vega.append(vega1)
    return vega

def strike(x, sig, T, t, r, K):
    strike = []
    for x in h: 
        strike1 = - np.exp( - r * (T - t)) * stats.norm.cdf(get_d_plus(x, sig, T, t, r, K))
        strike.append(strike1)
    return strike

def percent(x, sig, T, t, r, K):
    percent = []
    for x in h: 
        percent1 = K * np.exp( -r * (T - t)) * (T - t) * stats.norm.cdf(get_d_minus(x, sig, T, t, r, K))
        percent.append(percent1)
    return percent
```

```python
plt.plot(h, delta(x, sig, T, t, r, K), label='delta')
plt.plot(h, gamma(x, sig, T, t, r, K), label='gamma')
plt.plot(h, theta(x, sig, T, t, r, K), label='theta')
plt.plot(h, vega(x, sig, T, t, r, K), label='vega')
plt.plot(h, strike(x, sig, T, t, r, K), label='strike')
# plt.plot(h, percent(x, sig, T, t, r, K), label='percent')

plt.legend()
plt.grid()
```

```python
def get_xt(x, sig, T, t, r, K, N):
    Time = np.linspace(t, T, N)
    portfile = []
    st_last = x
    
    for i in Time:
        wt = np.random.normal(0, 1 / N , 1)[0]
        st_new = st_last * np.exp((r - sig ** 2 / 2) * 1 / N + sig * wt  )
        price_option = st_new * stats.norm.cdf(get_d_plus(st_new, sig, T, t, r, K)) - K * np.exp( - r * (T - i)) * stats.norm.cdf(get_d_minus(st_new, sig, T, t, r, K)) 
        portfile_cost = price_option - stats.norm.cdf(get_d_plus(st_new, sig, T, t, r, K)) * st_new
        portfile.append(portfile_cost)
        st_last = st_new
    return portfile

def get_xt_stab(x, sig, T, t, r, K, N):
    st = x
    xt_stab = []
    Time = np.linspace(t, T, N)
    Call_option = st * stats.norm.cdf(get_d_plus(st, sig, T, t, r, K)) - K * np.exp( - r * (T - t)) * stats.norm.cdf(get_d_minus(st, sig, T, t, r, K))    
    x0 = Call_option - stats.norm.cdf(get_d_plus(st, sig, T, t, r, K)) * st
    
    for i in Time:
        func = x0 * np.exp(r * i)
        xt_stab.append(func)
    return xt_stab
```

```python
N = 10000
Time = np.linspace(t, T, N)
plt.plot(Time, get_xt_stab(10, 0.1, 10, t, r, 20, N))
plt.plot(Time, get_xt(10, 0.1, 10, t, r, 20, N))
plt.grid()
```

```python
def put_theta_po_K(x, sig, T, t, r, K):
    for K in range(0, 10):
        theta = []
        for x in h: 
            theta1 = - x * sig * np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) * np.sqrt(T - t) / (np.sqrt(2 * np.pi) * 2) + r * K * np.exp( - r * (T - t) * stats.norm.cdf( - get_d_minus(x, sig, T, t, r, K)))
            theta.append(theta1)
        plt.plot(theta)
    plt.grid()
    return

def put_theta_po_sig(x, sig, T, t, r, K):
    sig_mas = np.linspace(0, 1, 10)  
    for sig in sig_mas:
        theta = []
        for x in h: 
            theta1 = - x * sig * np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) * np.sqrt(T - t) / (np.sqrt(2 * np.pi) * 2) + r * K * np.exp( - r * (T - t) * stats.norm.cdf( - get_d_minus(x, sig, T, t, r, K)))
            theta.append(theta1)
        plt.plot(theta)
    plt.grid()
    return 

def put_theta_po_r(x, sig, T, t, r, K):
    r_mas = np.linspace(0.01, 0.1, 10)
    for r in r_mas:
        theta = []
        for x in h: 
            theta1 = - x * sig * np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) * np.sqrt(T - t) / (np.sqrt(2 * np.pi) * 2) + r * K * np.exp( - r * (T - t) * stats.norm.cdf( - get_d_minus(x, sig, T, t, r, K)))
            theta.append(theta1)
        plt.plot(theta)
    plt.grid()
    return
```
