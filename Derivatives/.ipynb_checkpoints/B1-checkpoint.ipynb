```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.integrate as integrate
import scipy.stats as stats
```

```python
x = 1
r = 0.1
sig = 0.2
t = 0
T = 1
h = np.linspace(0, 100, 1001)
K = 20
```

```python
def get_d_plus(x, sig, T, t, r, strike):
    d_plus = (np.log(x / K) + (r + sig ** 2 / 2) * (T - t)) / (sig * np.sqrt(T - t))
    return d_plus

def get_d_minus(x, sig, T, t, r, strike):
    d_minus = (np.log(x / K) + (r - sig ** 2 / 2) * (T - t)) / (sig * np.sqrt(T - t))
    return d_minus

def theta(x, sig, T, t, r, K):
    theta = []
    for x in h: 
        theta1 = - x * sig * np.exp( - get_d_plus(x, sig, T, t, r, K) ** 2 / 2) * np.sqrt(T - t) / (np.sqrt(2 * np.pi) * 2) + r * K * np.exp( - r * (T - t) * stats.norm.cdf( - get_d_minus(x, sig, T, t, r, K)))
        theta.append(theta1)
    return theta


```

```python
plt.plot(h, theta(x, sig, T, t, r, K))
plt.grid()
```
