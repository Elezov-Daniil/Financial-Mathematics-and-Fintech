```python
import pandas as pd
from pandas import *
from openpyxl import *
import matplotlib.pyplot as plt 
from datetime import *
import numpy as np
```

```python
Open = pd.read_csv('Open.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
Close = pd.read_csv('Close.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
High = pd.read_csv('High.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
High.iloc[589, 1157] = High.iloc[589, 1157] * 10
High.iloc[590, 1157] = High.iloc[590, 1157] * 10
Low = pd.read_csv('Low.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
Volume = pd.read_csv('Volume.csv', index_col='Date').T
```

```python
def get_index_mistake(dataset):
    returns = (dataset.div(dataset.shift()) - 1).where((dataset.div(dataset.shift()) - 1) > 8, 0)
    date = returns.max(axis=1)
    idate = date.loc[date > 8]
    a = returns.iloc[idate.index - 1].max()
    table = a.loc[a > 8]
    c = returns.loc[idate.index, table.index]
    dateindex = []
    equityindex = []
    value = []
    for i in c.columns:
        for n in c.index:
            if c[i][n] != 0:
                dateindex.append(n-1)
                equityindex.append(i)
                value.append(c[i][n])
    return dateindex, equityindex, value
    

def get_filter(dataset):
    dataset = dataset.where(dataset != 0, 0.000000000001)
    a, b, c = get_index_mistake(dataset)
    for i in range(0, len(a)):
        if c[i] > 50:
            if dataset[b[i]][a[i]] * 5 < dataset[b[i]][a[i] - 1]:
                dataset[b[i]][a[i]] = dataset[b[i]][a[i]] * 100
            else:
                dataset[b[i]][a[i] + 1] = dataset[b[i]][a[i] + 1] / 100
        
        if c[i] < 10:
            if dataset[b[i]][a[i]] * 5 < dataset[b[i]][a[i] - 1]:
                dataset[b[i]][a[i]] = dataset[b[i]][a[i]] * 10
            else:
                dataset[b[i]][a[i] + 1] = dataset[b[i]][a[i] + 1] / 10
            
    dataset.columns = Volume.columns
    dataset.index = Volume.index
        
    return dataset.astype(np.float)

def test(dataset):
    test1 = (dataset.div(dataset.shift()) - 1)
    test2 = test1.max(axis=1)
    find_mistake = test2.loc[test2 > 8]
    return find_mistake
    
```

```python
Open = get_filter(Open)
Close = get_filter(Close)
High = get_filter(High)
Low = get_filter(Low)
```

```python
def neutralize(alpha):
    return alpha.sub(alpha.mean(axis=1), axis=0)

def normalize(alpha):
    return (alpha.div(alpha.abs().sum(axis=1), axis=0))

def truncate(alpha,max_pct):
    alpha = abs(alpha).where(alpha.abs() < max_pct, max_pct).mul(np.sign(alpha), axis=0)
    return alpha

def winsorise(alpha, num_std):
    p = alpha - alpha.mean()
    p = p.where(p.abs().sub(num_std*p.std(axis=1),axis=0)<0,np.sign(p).mul(num_std * p.std(axis=1),axis=0))
    alpha = p + alpha.mean()
    return alpha

def get_returns(alpha): 
    return (alpha.div(alpha.shift()) - 1)
    
def rank(alpha):
    return (alpha.rank(axis=1) - 1) / (alpha.shape[1] - 1)

def CutOutliers(alpha, n):
    a = rank(alpha)
    for i in range(0, n):
        a = a.where((a != i) & (a != a.shape[1] - i), 0)
    b = a.where(a == 0, 1)
    return alpha.mul(b, axis=0)

def CutMiddle(alpha, n):
    a = rank(alpha)
    for i in range(0, n):
        a = a.where((a != a.shape[1] / 2 + i) & (a != a.shape[1] / 2 - i), 0)
    # a = a.where(a != a.shape[1] / 2, 0)
    b = a.where(a == 0, 1)
    return alpha.mul(b, axis=0)

def apply_function(alpha, function):
    return list(map(function, alpha))

def decay(alpha, k):
    alpha_new = alpha
    for i in range(0, k):
        alpha_new = alpha_new.add(alpha.shift(1+i) * ((k - (1 + i)) / k))
    return alpha_new
```

```python
def get_turnover(alpha):
    turnover = abs(alpha.sub(alpha.shift(), axis=1)).sum(axis=1)
    # turnover1 = turnover.iloc[0:251].mean()
    # turnover2 = turnover.iloc[252:503].mean()
    # turnover3 = turnover.iloc[504:753].mean()
    # turnover4 = turnover.iloc[753:1005].mean()
    # turnover5 = turnover.iloc[1005:1257].mean()
    return turnover

def get_pnl(alpha):
    returns = get_returns(Close)
    pnl = alpha.shift().mul(returns).sum(axis=1)
    # pnl1 = pnl.iloc[0:251]
    # pnl2 = pnl.iloc[252:503]
    # pnl3 = pnl.iloc[504:753]
    # pnl4 = pnl.iloc[753:1005]
    # pnl5 = pnl.iloc[1005:1257]
    return pnl

def get_drawdown(pnl):
    pnl = pnl.cumsum()
    max_before_d = 0
    local_drawdown = 0
    max_drawdown = 0
    max_id = []
    max_value = []
    for idx, i in enumerate(pnl):
        if pnl[idx] >= max_before_d:
            max_before_d = pnl[idx]
            max_id.append(idx)
            max_value.append(i)
    max_id.append(idx)
    max_value.append(i)
    date_max_drawdown = 0
    for i in range(0, len(max_id)-1):
        local_drawdown = max_value[i] - pnl[max_id[i]:max_id[i+1]].min()
        if local_drawdown > max_drawdown:
            max_drawdown = local_drawdown
            date_max_drawdown = (pnl[max_id[i]:max_id[i+1]]).idxmin()
    return max_drawdown, date_max_drawdown

def get_sharpe(pnl):
    return pnl.mean() / pnl.std() * np.sqrt(len(pnl) - 1)
    
```

```python
def get_AlphaStats(alpha):
    pnl = get_pnl(alpha)
    pnl.index = pnl.index.astype(str).map(lambda x : datetime.strptime(x,'%Y-%m-%d'))
    pnl.cumsum().plot(label='alpha')
    plt.legend()
    plt.grid()

    
    years = [0, 252, 503, 753, 1005, 1257]
    AlphaStats = pd.DataFrame()
    for i in range(0, 5):
        turnover = get_turnover(alpha.iloc[years[i]:years[i+1]]).mean()
        sharpe = get_sharpe(pnl.iloc[years[i]:years[i+1]])
        pnl_local = get_pnl(alpha.iloc[years[i]:years[i+1]]).cumsum().max()
        drawdown, _ = get_drawdown(pnl.iloc[years[i]:years[i+1]])
        AlphaStats_years = pd.DataFrame([[pnl_local, sharpe, turnover, drawdown]], columns=['pnl', 'sharpe', 'turnover', 'drawdown'], index = [alpha.index[years[i]]])
        AlphaStats = AlphaStats.append(AlphaStats_years)
    return AlphaStats

def get_corr(pnl1, pnl2):
    df = pd.concat([pnl1, pnl2], axis=1)
    return df.corr(method='pearson')
```
