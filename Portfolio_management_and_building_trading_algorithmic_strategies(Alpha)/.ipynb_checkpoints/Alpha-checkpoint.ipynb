```python
import pandas as pd
from pandas import *
from openpyxl import *
import matplotlib.pyplot as plt 
from datetime import *
import numpy as np
```

```python
def normalize(alpha):
    return (alpha.div(alpha.abs().sum(axis=1), axis=0))
```

```python
def truncate(alpha,max_pct):
    alpha = abs(alpha).where(alpha.abs() < max_pct, max_pct).mul(np.sign(alpha), axis=0)
    return alpha
```

```python
def winsorise(alpha, num_std):
    p = alpha - alpha.mean()
    p = p.where(p.abs().sub(num_std*p.std(axis=1),axis=0)<0,np.sign(p).mul(num_std * p.std(axis=1),axis=0))
    alpha = p + alpha.mean()
    return alpha
```

```python
def neutralize(alpha):
    return alpha.sub(alpha.mean(axis=1), axis=0)
```

```python
def rebalance(alpha):
    counter = 0
    for m in alpha.index.drop(20100111):
        sum_pos = 0
        for i in alpha.columns:
            if alpha[i][m] >= 0:
                sum_pos += alpha[i][m]
            sum_neg = -(sum_pos - alpha.loc[m].sum())
        if sum_neg == 0 or sum_pos == 0:
            for i in alpha.columns:
                alpha.loc[m][i] = 0
        else:
            a = (sum_pos - alpha.loc[m].sum() / 2) / sum_pos
            b = (sum_neg - alpha.loc[m].sum() / 2) / sum_neg
            for i in alpha.columns:
                if alpha[i][m] >= 0:
                    alpha.loc[m][i] = alpha.loc[m][i] *  a
                else:
                    alpha.loc[m][i] = alpha.loc[m][i] *  b
        counter += 1
    return alpha
```
