```python
import pandas as pd
from pandas import *
from openpyxl import *
import matplotlib.pyplot as plt 
from datetime import *
import numpy as np
```

```python
Open = pd.read_csv('Open.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
Close = pd.read_csv('Close.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
High = pd.read_csv('High.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
High.iloc[589, 1157] = High.iloc[589, 1157] * 10
High.iloc[590, 1157] = High.iloc[590, 1157] * 10
Low = pd.read_csv('Low.csv').T.reset_index().drop(columns = ['index'],axis = 1).drop(labels = 0)
Volume = pd.read_csv('Volume.csv', index_col='Date').T
```

```python
def get_index_mistake(dataset):
    returns = (dataset.div(dataset.shift()) - 1).where((dataset.div(dataset.shift()) - 1) > 8, 0)
    date = returns.max(axis=1)
    idate = date.loc[date > 8]
    a = returns.iloc[idate.index - 1].max()
    table = a.loc[a > 8]
    c = returns.loc[idate.index, table.index]
    dateindex = []
    equityindex = []
    value = []
    for i in c.columns:
        for n in c.index:
            if c[i][n] != 0:
                dateindex.append(n-1)
                equityindex.append(i)
                value.append(c[i][n])
    return dateindex, equityindex, value
    

def get_filter(dataset):
    dataset = dataset.where(dataset != 0, 0.000000000001)
    a, b, c = get_index_mistake(dataset)
    for i in range(0, len(a)):
        if c[i] > 50:
            if dataset[b[i]][a[i]] * 5 < dataset[b[i]][a[i] - 1]:
                dataset[b[i]][a[i]] = dataset[b[i]][a[i]] * 100
            else:
                dataset[b[i]][a[i] + 1] = dataset[b[i]][a[i] + 1] / 100
        
        if c[i] < 10:
            if dataset[b[i]][a[i]] * 5 < dataset[b[i]][a[i] - 1]:
                dataset[b[i]][a[i]] = dataset[b[i]][a[i]] * 10
            else:
                dataset[b[i]][a[i] + 1] = dataset[b[i]][a[i] + 1] / 10
            
    dataset.columns = Volume.columns
    dataset.index = Volume.index
        
    return dataset

def test(dataset):
    test1 = (dataset.div(dataset.shift()) - 1)
    test2 = test1.max(axis=1)
    find_mistake = test2.loc[test2 > 8]
    return find_mistake
    
```

```python
Open = get_filter(Open)
Close = get_filter(Close)
High = get_filter(High)
Low = get_filter(Low)
```

```python
Open_test = test(Open)
Close_test = test(Close)
High_test = test(High)
Low_test = test(Low)
```

```python
def neutralize(alpha):
    return alpha.sub(alpha.mean(axis=1), axis=0)

def normalize(alpha):
    return (alpha.div(alpha.abs().sum(axis=1), axis=0))

def truncate(alpha,max_pct):
    alpha = abs(alpha).where(alpha.abs() < max_pct, max_pct).mul(np.sign(alpha), axis=0)
    return alpha

def winsorise(alpha, num_std):
    p = alpha - alpha.mean()
    p = p.where(p.abs().sub(num_std*p.std(axis=1),axis=0)<0,np.sign(p).mul(num_std * p.std(axis=1),axis=0))
    alpha = p + alpha.mean()
    return alpha

def returns_(alpha):
    returns = (alpha.div(alpha.shift()) - 1)
    return returns
    
def rank(alpha):
    alpha = alpha.rank(axis=1)
    return alpha

def CutOutliers(alpha, n):
    a = rank(alpha)
    for i in range(0, n):
        a = a.where((a != i) & (a != a.shape[1] - i), 0)
    b = a.where(a == 0, 1)
    return alpha.mul(b, axis=0)

def CutMiddle(alpha, n):
    a = rank(alpha)
    for i in range(0, n):
        a = a.where((a != a.shape[1] / 2 + i) & (a != a.shape[1] / 2 - i), 0)
    # a = a.where(a != a.shape[1] / 2, 0)
    b = a.where(a == 0, 1)
    return alpha.mul(b, axis=0)

def apply_function(alpha, function):
    alpha = list(map(function, alpha))
    return alpha
    

    
```
