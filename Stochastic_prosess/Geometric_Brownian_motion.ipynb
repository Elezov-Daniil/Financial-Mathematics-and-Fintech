```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
T = 5
N = 500
t = np.linspace(0, T, N + 1)
Delta = np.sqrt(T / N)
```

```python
def f(Delta, X):
    W = []
    W += [0]
    for i in range (1,len(X)):
        W += [ W [i -1]+ np.random.normal (0, 1, 1) * Delta ]
    return W

def Brownian_motion(T, N, t, Delta):
    X = np.linspace(0, T, N + 1)
    W = f(Delta, X)
    plt.figure ( figsize =(10, 7) )
    plt.grid()
    plt.plot(X, W, linewidth =2.0)
    plt.title('Wiener process')
    plt.xlabel("t")
    plt.ylabel("W")
    plt.show()
```

```python
def runif (n) :
    L =[]
    for i in range (n) :
        L += [2 * (np.random.uniform(0, 1) > 0.5) -1]
    print(L)
    return L

def function (x, n, S) :
    if x * n > 0:
        return S[int( x * n ) -1]
    else :
        return 0

def limit_of_a_random_walk (T, N, t):
    n = [10, 100, 1000]
    plt.figure ( figsize =(10, 7) )
    plt.grid ()
    for i in n:
        S = np.cumsum(runif(i))
        print(S)
        W = [function(x, i, S) / np.sqrt(i) for x in t]
        plt.plot (t , W , linewidth =2.0)

    plt.title ('Wiener process')
    plt.xlabel ("t")
    plt.ylabel ("W")
    plt.show ()
```

```python
def sum_W (t, T, Z):
    Sum = 0
    for i in range (len(Z)):
        Sum += Z[i] * (2 * np.sqrt(2 * T)) / ((2 * i + 1) * np.pi) * np.sin(((2 * i + 1) * np.pi * t) / (2 * T))
    return Sum

def L_squared_expansion(T, N, t):
    n = [10, 50, 100]
    Z = np.random.normal(0, 1, n)
    plt.figure(figsize = (10, 7))
    plt.grid()

    for k in n:
        W = []
        Z = np.random.normal(0, 1, k)
        for i in range(N + 1):
            W += [sum_W(t[i], T, Z)]
        plt.plot (t, W, linewidth =2.0)

    plt.title ('Wiener process')
    plt.xlabel ("t")
    plt.ylabel ("W")
```

```python
def geometric_Brownian_motion(T, N, t, Delta):
    r = 1
    sigma = 0.5
    x = 10
    W = [0]
    for i in range(N):
        W += [W[i-1] + np.random.normal(0, 1, 1) * Delta]
        S = [x * np.exp((r - sigma ** 2 / 2) * t[i] + sigma * W[i]) for i in range (len(W))]
    plt.figure ( figsize =(10, 7) )
    plt.grid()
    plt.plot(t, S, linewidth =2.0)
    plt.title('geometric Brownian motion')
    plt.xlabel("t")
    plt.ylabel("S")
    plt.show()
```

```python
def brownian_bridge(T, N, t, Delta):
    W = [0]
    for i in range(N):
        W +=[W[i-1] + np.random.normal(0, 1, 1) * np.sqrt ( Delta )]
    
    x =0
    y = -1
    BB =[x + W[i] - t[i] / T * ( W[N] - y + x ) for i in range ( N + 1)]
    plt.figure(figsize =(10, 7) )
    plt.grid()
    plt.plot (t, BB, linewidth =2.0)
    plt.scatter(t[0], x , color ="red",s =50)
    plt.scatter(t[len(t) -1] , y , color ="red", s =50)
    plt.title('Brownian bridge')
    plt.xlabel("t")
    plt.ylabel("BB")
    plt.show()

```
