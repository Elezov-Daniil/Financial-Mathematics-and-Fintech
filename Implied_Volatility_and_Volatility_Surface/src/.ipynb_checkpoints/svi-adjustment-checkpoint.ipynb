# SVI adjustment

```python
from dataclasses import dataclass
from math import *
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
@dataclass
class SVI:
    a: float
    b: float
    rho: float
    m: float
    sigma: float

    def jumpwing(
            self, t: float = 1) -> tuple[float, float, float, float, float]:
        """Returns parameters of the jump-wing parametrization.

        This parametrization depends on the time to expiration of the options
        constituting the volatility curve. See Gatheral's presentation (2004)
        for details and formulas.

        Args:
          t: Time to expiration of the options.

        Returns:
          Tuple  `(v, psi, p, c, v_tilde)` of the jump-wing parameteres.
        """
        w = (self.a +
             self.b*(-self.rho*self.m + np.sqrt(self.m**2+self.sigma**2)))
        v = w/t
        psi = self.b/np.sqrt(w)/2 * (
            -self.m/np.sqrt(self.m**2+self.sigma**2) + self.rho)
        p = self.b*(1-self.rho)/np.sqrt(w)
        c = self.b*(1+self.rho)/np.sqrt(w)
        v_tilde = (self.a + self.b*self.sigma*np.sqrt(1-self.rho**2)) / t
        return v, psi, p, c, v_tilde

    @classmethod
    def from_jumpwing(cls, v: float, psi: float, p: float, c: float,
                      v_tilde: float, t: float = 1):
        """Construct a class instance from jump-wing parameters."""
        w = v*t
        b = 0.5*np.sqrt(w)*(c+p)
        rho = 1 - 2*p/(c+p)
        beta = rho - 2*psi*np.sqrt(w)/b
        if np.abs(beta) > 1:
            raise ValueError(
                f"Smile is not convex: beta={beta}, but must be in [-1, 1].")
        elif beta == 0:
            m = 0
            sigma = (v-v_tilde)*t / (b*(1-np.sqrt(1-rho**2)))
        else:
            alpha = np.sign(beta) * np.sqrt(1/beta**2 - 1)
            m = (v-v_tilde)*t / (b*(-rho+np.sign(alpha)*np.sqrt(1+alpha**2) -
                                    alpha*np.sqrt(1-rho**2)))
            sigma = alpha*m
        a = v_tilde*t - b*sigma*np.sqrt(1-rho**2)
        return cls(a, b, rho, m, sigma)

    def __call__(self, x):
        """Returns the total implied variance `w(x)`."""
        return self.a + self.b*(self.rho*(x-self.m) +
                                np.sqrt((x-self.m)**2 + self.sigma**2))

    def durrleman_function(self, x):
        """Durrleman's function for verifying the convexity of a price surface.

        Args:
          x: Log-moneyness (scalar or array).

        Returns:
          The value of Durrleman's function. If `x` is an array, then an array
          of the same shape is returned.
        """
        # Total variance and its two derivatives
        w = self.__call__(x)
        wp = self.b*(self.rho + (x-self.m) / np.sqrt(
            (x-self.m)**2 + self.sigma**2))
        wpp = self.b*(1/np.sqrt((x-self.m)**2 + self.sigma**2) -
                      (x-self.m)**2/((x-self.m)**2 + self.sigma**2)**1.5)
        return (1-0.5*x*wp/w)**2 - 0.25*wp**2*(1/w+0.25) + 0.5*wpp
            
def vogt_example() -> SVI:
    """Returns an SVI parametrization not satisfying Durrleman's condition.

    See Gatheral, Jacquier "Arbitrage-free SVI volatility surfaces" (2014),
    Example 3.1 for details.
    """
    return SVI(a=-0.0410, b=0.1331, rho=0.3060, m=0.3586, sigma=0.4153)
```

## Vogt's example & Gatheral-Jacquier and Ferhati adjustments

Both adjustments give volatility curves quite far from the original one.

```python
svi_vogt = vogt_example()
svi_gatheral_crude = SVI.from_jumpwing(*svi_vogt.jumpwing()[:3], 0.3493158, 0.01548182, t=1)
svi_gatheral = SVI.from_jumpwing(*svi_vogt.jumpwing()[:3], 0.8564763, 0.0116249, t=1)
svi_ferhati = SVI(a=1e-5, b=0.08414, m=0.24957, rho=0.16962, sigma=0.1321)

X = np.linspace(-1.5, 1.5, 100)
plt.plot(X, np.sqrt(svi_vogt(X)), label="Vogt", color='k')
plt.plot(X, np.sqrt(svi_gatheral_crude(X)), label="Gatheral-Jacquier (crude)", color='b', linestyle='--', alpha=0.25)
plt.plot(X, np.sqrt(svi_gatheral(X)), label="Gatheral-Jacquier", color='b')
plt.plot(X, np.sqrt(svi_ferhati(X)), label="Ferhati", color='green')
plt.legend()
plt.xlabel("Log-moneyness")
plt.ylabel("Implied volatility")
plt.grid()
plt.show()

plt.plot(X, svi_vogt.durrleman_function(X), label="Vogt", color='k')
plt.plot(X, svi_gatheral_crude.durrleman_function(X), label="Gatheral-Jacquier (crude)", color='b', linestyle='--', alpha=0.25)
plt.plot(X, svi_gatheral.durrleman_function(X), label="Gatheral-Jacquier", color='b')
plt.plot(X, svi_ferhati.durrleman_function(X), label="Ferhati", color='green')
plt.legend()
plt.xlabel("Log-moneyness")
plt.ylabel("Durrleman's function")
plt.grid()
plt.show()
```

## Adjustment based on a correction of Durrleman's function

The condition for the absence of butterfly arbitrage is
$$
g(x) := \left(1-\frac{xw'(x)}{2w(x)}\right)^2 - \frac14 (w'(x))^2 \left(\frac1{w(x)}+\frac14\right) + \frac12 w''(x) \ge 0
$$
($x$ is the log-moneyness, $w(x)$ is the total variance). 

Let $g(x)$ be the corresponding function of an SVI curve. We correct the curve and construct $\tilde w(x)$ by solving
$$
\left\{
\begin{aligned}
&\frac12 \tilde w''(x) = g(x)^+ - \left(1-\frac{x\tilde w'(x)}{2\tilde w(x)}\right)^2 + \frac14 (\tilde w'(x))^2 \left(\frac1{\tilde w(x)}+\frac14\right),\\
&\tilde w(x_0) = w(x_0),\\
&\tilde w'(x_0) = w'(x_0),
\end{aligned}
\right.
$$
where $x_0$ is the left point of the curve.

By construction, Durrleman's function of $\tilde w(x)$ is equal to $\tilde g(x) = g(x)^+\ge 0$, hence the new curve is free of butterfly arbitrage.

```python
S = svi_vogt
X = np.linspace(-1.5, 1.5, 10000)

# new Durrleman's function
g = lambda x: max(S.durrleman_function(x), 0)
# also try - almost no difference
#x_critical = min(X[S.durrleman_function(X) < 0])
#g = lambda x: S.durrleman_function(x) if x < x_critical else 0


# RHS of the ODE
def f(y, x):
    w, dw = y
    return [dw, 2*g(x) + 0.5*(dw)**2*(1/w+0.25) - 2*(1-0.5*x*dw/w)**2]

# Initial conditions
y0 = [S(X[0]), (S(X[1])-S(X[0]))/(X[1]-X[0])]   # y0 = [w(x_0), w'(x_0)]

sol = odeint(f, y0, X)

w_new = sol[:, 0]
plt.plot(X, np.sqrt(S(X)), label="Vogt", color='k')
plt.plot(X, np.sqrt(svi_gatheral(X)), label="Gatheral-Jacquier", color='b')
plt.plot(X, np.sqrt(svi_ferhati(X)), label="Ferhati", color='green')
plt.plot(X, np.sqrt(w_new), label="New", color='r')
plt.legend()
plt.xlabel("Log-moneyness")
plt.ylabel("Implied volatility")
plt.grid()
plt.show()

def compute_g(X, W):
    wp = np.diff(W)/np.diff(X)
    wpp = np.diff(wp)/np.diff(X[1:])
    w = W[2:]
    wp = wp[1:]
    x = X[2:]
    return x, (1-0.5*x*wp/w)**2 - 0.25*(wp)**2*(1/w+0.25) + 0.5*wpp

plt.plot(X, S.durrleman_function(X), label="Vogt", color='k')
plt.plot(X, svi_gatheral.durrleman_function(X), label="Gatheral-Jacquier", color='b')
plt.plot(X, svi_ferhati.durrleman_function(X), label="Ferhati", color='green')
plt.plot(*compute_g(X, w_new), label="New", color='r')
plt.legend()
plt.xlabel("Log-moneyness")
plt.ylabel("Durrleman's function")
plt.grid()
plt.show()
```
