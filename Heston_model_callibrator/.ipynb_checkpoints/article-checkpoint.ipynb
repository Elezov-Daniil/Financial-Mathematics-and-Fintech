```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
U = 100
NU = 300
u, h = np.linspace(0, U, NU, retstep=True)
u = u + 0.5 * h
v = 0.15
theta = 0.15

rho = 0.5
k = 2.5
sig = 0.15

r = 0.1
s = 200.0
T = 1
K = 1000.0
NK = 5
NT = 8
Kmesh = np.linspace(0.5, 1.5, NK) * s
Tmesh = np.linspace(0.25, 4, NT)
C = np.zeros((NT, NK))

pay_off = []
```

```python
def get_phi(u, s, T, v, theta, rho, k, sig):
    x = np.log(s)
    du = np.sqrt((rho * sig * 1j * u - k) ** 2 + sig ** 2 * (1j * u + u ** 2))
    gu = (rho * sig * u * 1j - k + du) / (rho * sig * u * 1j - k - du)
    c1 = r * u * T * 1j + (k * theta / sig ** 2) * ((k - rho * sig * 1j * u - du) * T - (2 * np.log((1 - gu * np.exp(-du * T))/(1 - gu))))
    d1 = ((k - rho * sig * 1j * u - du) / sig ** 2) * (((1-np.exp(-du * T)) / (1 - gu * np.exp(-du * T))))
    phi = np.exp(c1 + v * d1 + u * 1j * x)

    return phi
```

```python
def get_phi_article(u, s, T, v, theta, rho,  k, sig):
    x = np.log(s)
    xi = k - sig * rho * 1j * u
    d = np.sqrt(xi ** 2 + sig ** 2 * ( 1j * u + u ** 2))
    D = np.log(d / v) + (k - d) * T / 2 - np.log(((d + xi) / 2 / v) + ((d - xi) * np.exp(-d * T) / 2 / v))
    A1 = (u ** 2 + 1j * u) * np.sinh(d * T / 2)
    A2 = d * np.cosh(d * T / 2) + xi * np.sinh(d * T / 2)
    A = A1 / A2
    log_phi = 1j * u * x + 1j * u * r * T - k * theta * rho * 1j * T * u / sig - A * v + 2 * k * theta / sig ** 2 * D
    phi = np.exp(log_phi)
    return phi
```

```python
u1, h1 = np.linspace(0, 4, 300, retstep=True)
s1 = 300
phi = (get_phi(u1, s1, T, v, theta, rho, k, sig)).real
phi_article = (get_phi_article(u1, s1, T, v, theta, rho,  k, sig)).real
plt.plot(u1, phi, '-', label='phi Heston', linewidth=1, color='red')
plt.plot(u1, phi_article, '_', label='phi article', linewidth=1)
plt.xlabel('u')
plt.ylabel('Re Phi')
plt.legend()
plt.grid()
```

```python
def get_phi_cherta(u, s, T, v, theta, rho, k, sig):
    phi_cherta = get_phi_article(u - 1j, s, T, v, theta, rho, k, sig) / get_phi_article(-1j, s, T, v, theta, rho, k, sig)
    return phi_cherta
```

```python
def get_F(u, s, T, K, v, theta, rho, k, sig):
    F = (np.exp(-u * 1j * np.log(K)) * get_phi_article(u, s, T, v, theta, rho, k, sig) / (1j * u)).real
    F_cherta = ((np.exp(-1j * u * np.log(K)) * get_phi_cherta(u, s, T, v, theta, rho, k, sig)) / (1j * u)).real
    return F, F_cherta   
```

```python
def get_phi_der(u, s, T, v, theta, rho, k, sig):
    x = np.log(s)
    xi = k - sig * rho * 1j * u
    d = np.sqrt(xi ** 2 + sig ** 2 * ( 1j * u + u ** 2))
    D = np.log(d / v) + (k - d) * T / 2 - np.log(((d + xi) / 2 / v) + ((d - xi) * np.exp(-d * T) / 2 / v))
    A1 = (u ** 2 + 1j * u) * np.sinh(d * T / 2)
    A2 = (d * np.cosh(d * T / 2) + xi * np.sinh(d * T / 2))/v
    A = A1 / A2
    log_phi = 1j * u * x + 1j * u * r * T - k * theta * rho * 1j * T * u / sig - A + 2 * k * theta / sig ** 2 * D
    phi = np.exp(log_phi)
    B = np.exp(D)
    d_rho = -xi * sig * 1j * u / d
    A2_rho = -sig  * 1j * u * (2 + xi * T) * (xi * np.cosh(d * T / 2) + d * np.sinh(d * T / 2)) / 2 / d / v
    B_rho = np.exp(k * T / 2) / v * (1 / A2 * d_rho - d / A2 ** 2 * A2_rho)
    A1_rho = - 1j * u * (u ** 2 + 1j * u) * T * xi * sig / 2 / d * np.cosh(d * T / 2)
    A_rho = A1_rho / A2 - A / A2 * A2_rho
    A_k = 1j / sig / u * A_rho
    B_k = 1j / sig / u * B_rho + B * T / 2
    d_sig = (rho / sig - 1 / xi) * d_rho + sig * u ** 2 / d
    A1_sig = (u ** 2 + 1j * u) * T / 2 * d_sig * np.cosh(d * T / 2)
    A2_sig = rho / sig * A2_rho - (2 + T * xi) / (v * T * xi * u * 1j) * A1_rho + sig * T * A1 / 2 / v
    A_sig = A1_sig / A2 - A / A2 * A2_sig
    hv = -A / v
    htheta = (2 * k) * D / sig ** 2 - 1j * k * rho * T * u / sig
    hrho = - A_rho + 2 * k * theta / sig ** 2 / d * (d_rho - d / A2 * A2_rho) - k * theta * T * 1j * u / sig
    hk = A_rho / sig / 1j / u + 2 * theta * D / sig ** 2 + 2 * k * theta * B_k / sig ** 2 / B - theta * rho * T * 1j * u / sig
    hsig = -A_sig - 4 * k * theta / sig ** 3 * D + 2 * k * theta / sig ** 2 / d * (d_sig - A2_sig * d / A2) + k * theta * rho * T * 1j * u / sig ** 2
    return phi, hv, htheta, hrho, hk, hsig
```

```python
def proverka_der(u, s, T, v, theta, rho,  k, sig):
    E = 1e-8
    phi1 = get_phi_article(u, s, T, v + E, theta, rho,  k, sig)
    phi2 = get_phi_article(u, s, T, v - E, theta, rho,  k, sig)
    proverka_v = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = get_phi_article(u, s, T, v, theta + E, rho,  k, sig)
    phi2 = get_phi_article(u, s, T, v, theta - E, rho,  k, sig)
    proverka_theta = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = get_phi_article(u, s, T, v, theta, rho + E,  k, sig)
    phi2 = get_phi_article(u, s, T, v, theta, rho - E,  k, sig)
    proverka_rho = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = get_phi_article(u, s, T, v, theta, rho,  k + E, sig)
    phi2 = get_phi_article(u, s, T, v, theta, rho,  k - E, sig)
    proverka_k = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = get_phi_article(u, s, T, v, theta, rho,  k, sig + E)
    phi2 = get_phi_article(u, s, T, v, theta, rho,  k, sig - E)
    proverka_sig = (np.log(phi1) - np.log(phi2)) / 2 / E
    return proverka_v, proverka_theta, proverka_rho, proverka_k, proverka_sig
```

```python
def get_phi_der_cherta(u, s, T, v, theta, rho, k, sig):
    phi_ch, hv_ch, htheta_ch, hrho_ch, hk_ch, hsig_ch = get_phi_der(u - 1j, s, T, v, theta, rho, k, sig)
    phi_temp = get_phi_article(-1j, s, T, v, theta, rho, k, sig)
    phi_ch = phi_ch / phi_temp
    return phi_ch, hv_ch, htheta_ch, hrho_ch, hk_ch, hsig_ch
```

```python
def get_p_der(u, s, T, K, v, theta, rho,  k, sig, h):
    phi, hv, htheta, hrho, hk, hsig = get_phi_der(u, s, T, v, theta, rho,  k, sig) 
    
    
    F = (np.exp(-u * 1j * np.log(K)) * phi / (1j * u)).real
    P = 0.5 + 1 / np.pi * np.sum(F * h)
    F1 = F
    
    F = (np.exp(-u * 1j * np.log(K)) * phi *  hv / (1j * u)).real
    Pv = 1 / np.pi * np.sum(F * h)

    F = (np.exp(-u * 1j * np.log(K)) * phi * htheta / (1j * u)).real
    Ptheta = 1 / np.pi * np.sum(F * h)

    F = (np.exp(-u * 1j * np.log(K)) * phi * hrho / (1j * u)).real
    Prho = 1 / np.pi * np.sum(F * h)

    F = (np.exp(-u * 1j * np.log(K)) * phi * hk / (1j * u)).real
    Pk = 1 / np.pi * np.sum(F * h)

    F = (np.exp(-u * 1j * np.log(K)) * phi * hsig / (1j * u)).real
    Psig = 1 / np.pi * np.sum(F * h)
    
    return P, Pv, Ptheta, Prho, Pk, Psig
    
```

```python
def get_p_der_cherta(u, s, T, K, v, theta, rho, k, sig, h): 
    phi_ch, hv_ch, htheta_ch, hrho_ch, hk_ch, hsig_ch = get_phi_der_cherta(u, s, T, v, theta, rho, k, sig)
    
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch / (1j * u)).real
    P_cherta = 0.5 + 1 / np.pi * np.sum(F_cherta * h)
    F1 = F_cherta          
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch * hv_ch / (1j * u)).real
    Pv_ch = 1 / np.pi * np.sum(F_cherta * h)
    
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch * htheta_ch / (1j * u)).real
    Ptheta_ch = 1 / np.pi * np.sum(F_cherta * h)
                
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch * hrho_ch / (1j * u)).real
    Prho_ch = 1 / np.pi * np.sum(F_cherta * h)
                
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch * hk_ch / (1j * u)).real
    Pk_ch = 1 / np.pi * np.sum(F_cherta * h)
    
    F_cherta = (np.exp(-1j * u * np.log(K)) * phi_ch * hsig_ch / (1j * u)).real
    Psig_ch = 1 / np.pi * np.sum(F_cherta * h)
    
    return P_cherta, Pv_ch, Ptheta_ch, Prho_ch, Pk_ch, Psig_ch
```

```python
def get_price_option_C(u, s, T, K, r, v, theta, rho,  k, sig, h):
    P_cherta, *_ = get_p_der_cherta(u, s, T, K, v, theta, rho,  k, sig, h)
    P, *_ = get_p_der(u, s, T, K, v, theta, rho,  k, sig, h)
    C = s * P_cherta - np.exp(-r * T) * K * P
    return C
```

```python
def get_price_option_der(u, s, T, K, r, v, theta, rho,  k, sig, h):
    P_cherta, Pv_ch, Ptheta_ch, Prho_ch, Pk_ch, Psig_ch = get_p_der_cherta(u, s, T, K, v, theta, rho, k, sig, h)
    P, Pv, Ptheta, Prho, Pk, Psig = get_p_der(u, s, T, K, v, theta, rho,  k, sig, h)
    C = s * P_cherta - np.exp(-r * T) * K * P
    Cv = s * Pv_ch - np.exp(-r * T) * K * Pv
    Ctheta = s * Ptheta_ch - np.exp(-r * T) * K * Ptheta
    Crho = s * Prho_ch - np.exp(-r * T) * K * Prho
    Ck = s * Pk_ch - np.exp(-r * T) * K * Pk
    Csig = s * Psig_ch - np.exp(-r * T) * K * Psig
    return C, Cv, Ctheta, Crho, Ck, Csig
```

```python
def proverka_c_der(u, s, T, K, r, v, theta, rho,  k, sig, h):
    E = 1e-8
    C1 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k, sig, h)
    C2 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k, sig, h)
    proverka_C = C1
    
    C1 = get_price_option_C(u, s, T, K, r, v + E, theta, rho,  k, sig, h)
    C2 = get_price_option_C(u, s, T, K, r, v - E, theta, rho,  k, sig, h)
    proverka_Cv = (C1 - C2) / 2 / E
    
    C1 = get_price_option_C(u, s, T, K, r, v, theta + E, rho,  k, sig, h)
    C2 = get_price_option_C(u, s, T, K, r, v, theta - E, rho,  k, sig, h)
    proverka_Ctheta = (C1 - C2) / 2 / E
    
    C1 = get_price_option_C(u, s, T, K, r, v, theta, rho + E,  k, sig, h)
    C2 = get_price_option_C(u, s, T, K, r, v, theta, rho - E,  k, sig, h)
    proverka_Crho = (C1 - C2) / 2 / E
    
    C1 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k + E, sig, h)
    C2 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k - E, sig, h)
    proverka_Ck = (C1 - C2) / 2 / E
    
    C1 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k, sig + E, h)
    C2 = get_price_option_C(u, s, T, K, r, v, theta, rho,  k, sig - E, h)
    proverka_Csig = (C1 - C2) / 2 / E
    
    return proverka_C, proverka_Cv, proverka_Ctheta, proverka_Crho, proverka_Ck, proverka_Csig
```

```python
T = 1.0
C_hat = np.zeros((NK, ))
for ik, Ki in enumerate(Kmesh):
    F, F_cherta = get_F(u, s, T, Ki, v, theta, rho, k, sig)
    P = 0.5 + 1 / np.pi * np.sum(F * h)
    P_cherta = 0.5 + 1 / np.pi * np.sum(F_cherta * h)
    price = s * P_cherta - np.exp(-r * T) * Ki * P
    C_hat[ik] = price
```

```python
v0 = 0.35
theta0 = 0.25
rho0 = -0.5
k0 = 1.5
sig0 = 1.0

C_hat = []
T_hat = []
K_hat = []
for it, Ti in enumerate(Tmesh):
    for ik, Ki in enumerate(Kmesh):
        C_hat.append(get_price_option_C(u, s, Ti, Ki, r, v0, theta0, rho0,  k0, sig0, h))
        T_hat.append(Ti)
        K_hat.append(Ki)
C_hat = np.asarray(C_hat)
T_hat = np.asarray(T_hat)
K_hat = np.asarray(K_hat)
```

```python
def get_residuals(u, s, T_hat, K_hat, r, v, theta, rho,  k, sig, h, C_hat):
    J = np.zeros((5, len(C_hat)))
    res = np.zeros((len(C_hat)))
    
    for i in range(len(C_hat)):
        C, Cv, Ctheta, Crho, Ck, Csig = get_price_option_der(u, s, T_hat[i], K_hat[i], r, v, theta, rho,  k, sig, h)
        res[i] = C - C_hat[i]
        J[0, i] = Cv 
        J[1, i] = Ctheta
        J[2, i] = Crho
        J[3, i] = Ck
        J[4, i] = Csig
        
    return J, res
```

```python
def proj(v, theta, rho, k, sig):
    E = 1e-4
    v = max(v, E)
    theta = max(theta, E)
    rho = max(rho, -1 + E)
    rho = min(rho, 1 - E)
    k = max(k, E)
    sig = max(sig, E)
    return v, theta, rho, k ,sig
```

```python
def grad_desc(u, s, T, Kmesh, r, v, theta, rho,  k, sig, h, C_hat):
    step = 0.0001
    Niter = 100
    fs = []
    for i in range(Niter):
        v1 = v
        theta1 = theta
        rho1 = rho
        k1 = k
        sig1 = sig
        J, res = get_residuals(u, s, T, Kmesh, r, v1, theta1, rho1,  k1, sig1, h, C_hat)
        grad = J @ res
        v = v1 - step * grad[0]
        theta = theta1 - step * grad[1]
        rho = rho1 - step * grad[2]
        k = k1 - step * grad[3]
        sig = sig1 - step * grad[4]
        fs.append(np.sum(res ** 2))
        v, theta, rho, k, sig = proj(v, theta, rho, k, sig)
    return fs, v, theta, rho, k, sig
```

```python
def get_random_params():
    v = 0.1 + np.random.rand()
    theta = 0.1 + np.random.rand()
    rho = -0.9 + 1.8 * np.random.rand()
    k = 1 + 1.5 * np.random.rand()
    sig = 0.1 + np.random.rand()
    return v, theta, rho, k, sig
```

```python
def grad(u, s, T_hat, K_hat, r, v, theta, rho,  k, sig, h, C_hat):
    step = 1 / 0.1
    Niter = 200
    J, res = get_residuals(u, s, T_hat, K_hat, r, v, theta, rho,  k, sig, h, C_hat)
    fs = [np.sum(res ** 2)]
    prev_fs = np.sum(res ** 2)
    vk, thetak, rhok, kk, sigk = [], [], [], [], []
    
    for i in range(Niter):
        I = np.eye(5)
        I = np.diag(np.diag( J @ J.T ) )
        A = step * I + J @ np.transpose(J) 
        dtheta = np.linalg.solve(A, J @ res)
        v1 = v - dtheta[0]
        theta1 = theta - dtheta[1]
        rho1 = rho - dtheta[2]
        k1 = k - dtheta[3]
        sig1 = sig - dtheta[4]
        v1, theta1, rho1, k1, sig1 = proj(v1, theta1, rho1, k1, sig1)
        
        J1, res1 = get_residuals(u, s, T_hat, K_hat, r, v1, theta1, rho1,  k1, sig1, h, C_hat)
    
        cur_fs = np.sum(res1 ** 2)
        
        if cur_fs < prev_fs:
            step /= 1.5
            prev_fs = cur_fs
            fs.append(cur_fs)
            J = J1
            res = res1
            v, theta, rho, k, sig = v1, theta1, rho1, k1, sig1
            vk.append(v1)
            thetak.append(theta1)
            rhok.append(rho)
            kk.append(k1)
            sigk.append(sig1)
            
        else:
            step *= 1.5
        if cur_fs < 1e-15:
            break
            
    vk = np.asarray(vk)
    thetak = np.asarray(thetak)
    rhok = np.asarray(rhok)
    kk = np.asarray(kk)
    sigk = np.asarray(sigk)
    return fs, v, theta, rho, k, sig, vk, thetak, rhok, kk, sigk
```

```python
v1, theta1, rho1, k1, sig1 = get_random_params()
fs, v, theta, rho, k, sig, vk, thetak, rhok, kk, sigk = grad(u, s, T_hat, K_hat, r, v1, theta1, rho1,  k1, sig1, h, C_hat)
```
