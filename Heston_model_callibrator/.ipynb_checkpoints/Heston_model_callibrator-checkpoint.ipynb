```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
U = 10
NU = 1000
u, h = np.linspace(0, U, NU, retstep=True)
v = 0.1
theta = 0.1

rho = 0.5
k = 2.5
sig = 0.1

r = 0.1
s = 500.0
time = 4.0
K = 1000.0
NK = 1000
NT = 11
Kmesh = np.linspace(1000, 1.5, NK)
Tmesh = np.linspace(0.25, 2, NT)
C = np.zeros((NT, NK))
pay_off = []
```

```python
def get_phi(u, s, time, v, theta, rho, k, sig):
    x = np.log(s)
    du = np.sqrt((rho * sig * 1j * u - k) ** 2 + sig ** 2 * (1j * u + u ** 2))
    gu = (rho * sig * u * 1j - k + du) / (rho * sig * u * 1j - k - du)
    c1 = r * u * time * 1j + (k * theta / sig ** 2) * ((k - rho * sig * 1j * u - du) * time - (2 * np.log((1 - gu * np.exp(-du * time))/(1 - gu))))
    d1 = ((k - rho * sig * 1j * u - du) / sig ** 2) * ((1-np.exp(-du * time) / (1 - gu * np.exp(-du * time))))
    phi = np.exp(c1 + v * d1 + u * 1j * x)

    return phi
```

```python
def phi_article(u, s, time, v, theta, rho,  k, sig, h):
    x = np.log(s)
    xi = k - sig * rho * 1j * u
    d = np.sqrt(xi ** 2 + sig ** 2 * ( 1j * u + u ** 2))
    D = np.log(d / v) + (k - d) * time / 2 - np.log(((d + xi) / 2 / v) + ((d - xi) * np.exp(-d * time) / 2 / v))
    A1 = (u ** 2 + 1j * u) * np.sinh(d * time / 2)
    A2 = d * np.cosh(d * time / 2) + xi * np.sinh(d * time / 2)
    A = A1 / A2
    log_phi = 1j * u * x + 1j * u * r * time - k * theta * rho * 1j * time * u / sig - A * v + 2 * k * theta / sig ** 2 * D
    phi = np.exp(log_phi)
    return phi
   

    
```

```python
def phi_der(u, s, time, v, theta, rho,  k, sig, h):
    x = np.log(s)
    xi = k - sig * rho * 1j * u
    d = np.sqrt(xi ** 2 + sig ** 2 * ( 1j * u + u ** 2))
    D = np.log(d / v) + (k - d) * time / 2 - np.log(((d + xi) / 2 / v) + ((d - xi) * np.exp(-d * time) / 2 / v))
    A1 = (u ** 2 + 1j * u) * np.sinh(d * time / 2)
    A2 = (d * np.cosh(d * time / 2) + xi * np.sinh(d * time / 2))/v
    A = A1 / A2
    log_phi = 1j * u * x + 1j * u * r * time - k * theta * rho * 1j * time * u / sig - A + 2 * k * theta / sig ** 2 * D
    phi = np.exp(log_phi)
    B = np.exp(D)
    d_rho = -xi * sig * 1j * u / d
    A2_rho = -sig  * 1j * u * (2 + xi * time) * (xi * np.cosh(d * time / 2) + d * np.sinh(d * time / 2)) / 2 / d / v
    B_rho = np.exp(k * time / 2) / v * (1 / A2 * d_rho - d / A2 ** 2 * A2_rho)
    A1_rho = - 1j * u * (u ** 2 + 1j * u) * time * xi * sig / 2 / d * np.cosh(d * time / 2)
    A_rho = A1_rho / A2 - A / A2 * A2_rho
    A_k = 1j / sig / u * A_rho
    B_k = 1j / sig / u * B_rho + B * time / 2
    d_sig = (rho / sig - 1 / xi) * d_rho + sig * u ** 2 / d
    A1_sig = (u ** 2 + 1j * u) * time / 2 * d_sig * np.cosh(d * time / 2)
    A2_sig = rho / sig * A2_rho - (2 + time * xi) / (v * time * xi * u * 1j) * A1_rho + sig * time * A1 / 2 / v
    A_sig = A1_sig / A2 - A / A2 * A2_sig
    hv = -A / v
    htheta = (2 * k) * D / sig ** 2 - 1j * k * rho * time * u / sig
    hrho = - A_rho + 2 * k * theta / sig ** 2 / d * (d_rho - d / A2 * A2_rho) - k * theta * time * 1j * u / sig
    hk = A_rho / sig / 1j / u + 2 * theta * D / sig ** 2 + 2 * k * theta * B_k / sig ** 2 / B - theta * rho * time * 1j * u / sig
    hsig = -A_sig - 4 * k * theta / sig ** 3 * D + 2 * k * theta / sig ** 2 / d * (d_sig - A2_sig * d / A2) + k * theta * rho * time * 1j * u / sig ** 2
    return phi, hv, htheta, hrho, hk, hsig
```

```python
def proverka_der(u, s, time, v, theta, rho,  k, sig, h):
    E = 1e-8
    phi1 = phi_article(u, s, time, v + E, theta, rho,  k, sig, h)
    phi2 = phi_article(u, s, time, v - E, theta, rho,  k, sig, h)
    proverka_v = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = phi_article(u, s, time, v, theta + E, rho,  k, sig, h)
    phi2 = phi_article(u, s, time, v, theta - E, rho,  k, sig, h)
    proverka_theta = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = phi_article(u, s, time, v, theta, rho + E,  k, sig, h)
    phi2 = phi_article(u, s, time, v, theta, rho - E,  k, sig, h)
    proverka_rho = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = phi_article(u, s, time, v, theta, rho,  k + E, sig, h)
    phi2 = phi_article(u, s, time, v, theta, rho,  k - E, sig, h)
    proverka_k = (np.log(phi1) - np.log(phi2)) / 2 / E
    
    phi1 = phi_article(u, s, time, v, theta, rho,  k, sig + E, h)
    phi2 = phi_article(u, s, time, v, theta, rho,  k, sig - E, h)
    proverka_sig = (np.log(phi1) - np.log(phi2)) / 2 / E
    return proverka_v, proverka_theta, proverka_rho, proverka_k, proverka_sig
```

```python
def get_phi_cherta(u, s, time, v, theta, rho, k, sig):
    phi_cherta = get_phi(u - 1j, s, time, v, theta, rho, k, sig) / get_phi(-1j, s, time, v, theta, rho, k, sig)
    
    return phi_cherta
```

```python
def get_F(u, s, time, v, theta, rho, k, sig, K):
    F = (np.exp(-u * 1j * np.log(K)) * get_phi(u, s, time, v, theta, rho, k, sig) / (1j * u)).real
    F_cherta = ((np.exp(-1j * u * np.log(K)) * get_phi_cherta(u, s, time, v, theta, rho, k, sig)) / (1j * u)).real
    return F, F_cherta
    
    
```

```python
F, F_cherta = get_F(u + 0.5 * h, s, time, v, theta, rho, k, sig, K)
P = 0.5 + 1 / np.pi * np.sum(F * h)
P_cherta = 0.5 + 1 / np.pi * np.sum(F_cherta * h)
```

```python
def price_option(s, P_cherta, r, time, K, P):
    price = s * P_cherta - np.exp(-r * time) * K * P
    return price
```

```python
for it, t in enumerate(Tmesh):
    for ik, i in enumerate(Kmesh): 
        F, F_cherta = get_F(u + 0.5 * h, s, t, v, theta, rho, k, sig, i)
        P = 0.5 + 1 / np.pi * np.sum(F * h)
        P_cherta = 0.5 + 1 / np.pi * np.sum(F_cherta * h)
        price = price_option(s, P_cherta, r, t, i, P)
        C[it, ik] = price
```

```python
phi_old = get_phi(u, s, time, v, theta, rho, k, sig)
```

```python
phi_new = phi_article(u, s, time, v, theta, rho,  k, sig, h)
```
