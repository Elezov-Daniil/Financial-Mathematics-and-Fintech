```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
T = 1
K = 100
r = 0.1
sigma = 0.1
sigma_ = 0.1
S = 100
t = 0
M = 100
N = 1000
```

```python
def value_call_option_Monte_Carlo(S, K, r, sigma, T, N):
    z = np.random.standard_normal(N)
    ST = S * np.exp((r - sigma ** 2 / 2) * T + sigma * np.sqrt(T) * z)
    payoff = np.maximum(ST - K, 0)
    return np.exp(-r * T) * np.mean(payoff)

def value_call_option_BSM(S, K, r, sigma, T, t):
    d1 = 1/(sigma*pow(T-t, 0.5))*(np.log(S/K)+(r + (sigma**2)/2)*(T-t))
    d2 = 1/(sigma*pow(T-t, 0.5))*(np.log(S/K)+(r - (sigma**2)/2)*(T-t))
    c = S * norm.cdf(d1) - K * np.exp(-1*r*(T-t))*norm.cdf(d2)
    return c

def value_asiatic_option_Monte_Carlo(S0, K, r, sigma, T, N, M):
    S = np.zeros((N, M + 1))
    z = norm.rvs(size=(N, M))
    sr = np.zeros(N)
    c = np.zeros(N)
    csum = 0
    for j in range(N):
        S[j, 0] = S0
        sumsr = 0
        for i in range(M):
            S[j, i + 1] = S[j, i] * np.exp((r - sigma ** 2 / 2) * (T / M) + sigma * np.sqrt(T / M) * z[j, i])
            sumsr +=  S[j, i+1]
            sr[j] = sumsr/M
            if (sr[j] - K) > 0:
                c[j] = (sr[j] - K) * np.exp(-r * T)
            else:
                c[j] = 0
        csum += c[j]
    return csum / N
```

```python
print('Value call option by Monte Carlo =', value_call_option_Monte_Carlo(S, K, r, sigma, T, N))
print('Value call option by BSM =', value_call_option_BSM(S, K, r, sigma, T, t))
print('Value call asiatic option by Monte Carlo =', value_asiatic_option_Monte_Carlo(S, K, r, sigma, T, N, 1))
```

```python
def explicit_scheme(S, K, M, N, T, t, r, sigma, sigma_):
    tau, dt = np.linspace(sigma ** 2 * T / 2, t, N, retstep=True)
    xt, dx = np.linspace(np.log(S / 4 / K), np.log(S*2 / K), M, retstep=True)
    u = pd.DataFrame(np.zeros((N, M)),columns=xt, index = tau)
    D = 2 * r / (sigma ** 2)
    for i in range(0, M):
        if (1 - np.exp(-xt[i])) > 0:
            u.iloc[N-1][i] = 1 - np.exp(-xt[i]) 
    for j in range(0, N):
        u.iloc[j, M-1] = 1 - np.exp( - D * tau[j] - xt[M-1])

    for j in range(N-1, 0, -1):
        for i in range(1, M - 1):
            u.iloc[j - 1, i] = (sigma_ ** 2 / sigma ** 2 * ((u.iloc[j, i - 1] - 2 * u.iloc[j, i] + u.iloc[j, i + 1]) / dx  ** 2\
                                                                            + (u.iloc[j, i + 1] - u.iloc[j, i - 1]) / dx / 2)\
                                                                            + D * (u.iloc[j, i + 1]- u.iloc[j, i - 1]) / dx / 2) * abs(dt) + u.iloc[j, i]
    return u, xt, dx, dt
```

```python
u, xt, dx, dt = explicit_scheme(S, K, M, N, T, t, r, sigma, sigma_)
vt = np.linspace(0, 0, len(u.iloc[0]))
st = np.linspace(0, 0, len(u.iloc[0]))
for i in range(0, len(u.iloc[0])):
    vt[i] = K * np.exp(xt[i]) * u.iloc[0, i]
    st[i] = K * np.exp(xt[i])
c = [value_call_option_BSM(st[i], K, r, sigma, T, t) for i in range(len(xt))]
print('Relative error =', (vt[int(np.round(2 * M / 3))] - c[int(np.round(2 * M / 3))])/c[int(np.round(2 * M / 3))])
print('Analytical error =', dx ** 2 + dt)
plt.plot(st, vt, label='explicits scheme')
plt.plot(st, c, label='BSM')
plt.xlabel("Value asset")
plt.ylabel("Value call option")
plt.grid()
plt.legend()
plt.show()
```

```python
def implicit_scheme(S, K, M, N, T, t, r, sigma, sigma_):
    tau, dt = np.linspace(sigma ** 2 * T / 2, t, N, retstep=True)
    xt, dx = np.linspace(np.log(S / 4 / K), np.log(S*2 / K), M, retstep=True)
    a = pd.DataFrame(np.zeros((N, M)),columns=xt, index = tau)
    D = 2 * r / (sigma ** 2)
    for i in range(0, M):
        if (1 - np.exp(-xt[i])) > 0:
            a.iloc[N-1][i] = 1 - np.exp(-xt[i]) 
    for j in range(0, N):
        a.iloc[j, M-1] = 1 - np.exp( - D * tau[j] - xt[M-1])
    sig = sigma_ ** 2 / sigma ** 2
    gamma = (- 2 * sig + dx * sig + D * dx) * abs(dt) / (2 * dx ** 2)
    alpha = (dx ** 2 + abs(dt) * sig * 2) / dx ** 2
    beta = (2 * sig + dx * sig + D * dx) * - abs(dt) / (2 * dx ** 2)

    for j in range(0, N-1):
        b = np.linspace(0,0, M)
        for i in range(0,M):
            b[i] = a.iloc[N-j-1, i]
        b[1] = a.iloc[N - j - 1, 1] - gamma * a.iloc[N - j - 2, 0] 
        b[M-2] = a.iloc[N - j - 1, M-2] - beta * a.iloc[N -j - 2, M-1]

        A = np.linspace(0,0, M)
        B = np.linspace(0,0, M)
        A[2] = - beta / alpha
        B[2] = b[1] / alpha
        for i in range(2, M-2):
            A[i+1] = - beta / (gamma * A[i] + alpha)
            B[i+1] = (b[i] - gamma * B[i]) / (gamma * A[i] + alpha)

        a.iloc[N - j - 2, M-2] = - (B[M-2] - b[M-2] / gamma) / (A[M-2] + alpha / gamma)
        for i in range(M-3, 0, -1):
            a.iloc[N-j - 2, i] = A[i+1] *a.iloc[N-j-2, i + 1] + B[i+1]
    return a, xt, dx, dt
```

```python
u, xt, dx, dt = implicit_scheme(S, K, M, N, T, t, r, sigma, sigma_)
vt = np.linspace(0, 0, len(u.iloc[0]))
st = np.linspace(0, 0, len(u.iloc[0]))
for i in range(0, len(u.iloc[0])):
    vt[i] = K * np.exp(xt[i]) * u.iloc[0, i]
    st[i] = K * np.exp(xt[i])
c = [value_call_option_BSM(st[i], K, r, sigma, T, t) for i in range(len(xt))]
print('Relative error =', (vt[int(np.round(2 * M / 3))] - c[int(np.round(2 * M / 3))])/c[int(np.round(2 * M / 3))])
print('Analytical error =', dx ** 2 + dt)
plt.plot(st, vt, label='implicit scheme')
plt.plot(st, c, label='BSM')
plt.xlabel("Value asset")
plt.ylabel("Value call option")
plt.grid()
plt.legend()
plt.show()
```

```python
def Krank_Nikolson_scheme(S, K, M, N, T, t, r, sigma, sigma_):
    tau, dt = np.linspace(sigma ** 2 * T / 2, t, N, retstep=True)
    xt, dx = np.linspace(np.log(S / 4 / K), np.log(S*2 / K), M, retstep=True)
    dt = abs(dt)
    a = pd.DataFrame(np.zeros((N, M)),columns=xt, index = tau)
    D = 2 * r / (sigma ** 2)
    for i in range(0, M):
        if (1 - np.exp(-xt[i])) > 0:
            a.iloc[N-1][i] = 1 - np.exp(-xt[i]) 
    for j in range(0, N):
        a.iloc[j, M-1] = 1 - np.exp( - D * tau[j] - xt[M-1])
    sig = sigma_ ** 2 / sigma ** 2
    gamma = - dt * sig /2 / dx ** 2  +  dt / 4 / dx * (sig + D)
    alpha = dt / dx ** 2 * sig + 1
    beta =  - (dt * sig /2 / dx ** 2  +  dt / 4 / dx * (sig + D))
    for j in range(0, N-1):
        b = np.linspace(0,0, M)
        for i in range(1,M - 1):
            b[i] = a.iloc[N-j-1, i] + 1 / 2 / dx ** 2 * dt * sig * (a.iloc[N-j-1, i - 1] - 2 * a.iloc[N-j-1, i] + a.iloc[N-j-1, i + 1]) + 1 / 4 / dx * dt * (sig + D) * ( - a.iloc[N-j-1, i - 1] + a.iloc[N-j-1, i + 1])
        b[1] = a.iloc[N - j - 1, 1] - gamma * a.iloc[N - j - 2, 0] 
        b[M-2] = a.iloc[N - j - 1, M-2] - beta * a.iloc[N -j - 2, M-1]


        A = np.linspace(0,0, M)
        B = np.linspace(0,0, M)
        A[2] = - beta / alpha
        B[2] = b[1] / alpha
        for i in range(2, M-2):
            A[i+1] = - beta / (gamma * A[i] + alpha)
            B[i+1] = (b[i] - gamma * B[i]) / (gamma * A[i] + alpha)

        a.iloc[N - j - 2, M-2] = - (B[M-2] - b[M-2] / gamma) / (A[M-2] + alpha / gamma)
        for i in range(M-3, 0, -1):
            a.iloc[N-j - 2, i] = A[i+1] * a.iloc[N-j-2, i + 1] + B[i+1]
    return a, xt, dx, dt
```

```python
M = 500
N = 500
u, xt, dx, dt = Krank_Nikolson_scheme(S, K, M, N, T, t, r, sigma, sigma_)
vt = np.linspace(0, 0, len(u.iloc[0]))
st = np.linspace(0, 0, len(u.iloc[0]))
for i in range(0, len(u.iloc[0])):
    vt[i] = K * np.exp(xt[i]) * u.iloc[0, i]
    st[i] = K * np.exp(xt[i])
c = [value_call_option_BSM(st[i], K, r, sigma, T, t) for i in range(len(xt))]
print('Relative error =', (vt[int(np.round(2 * M / 3))] - c[int(np.round(2 * M / 3))])/c[int(np.round(2 * M / 3))])
print('Analytical error =', dx ** 2 + dt ** 2)
plt.plot(st, vt, label='Krank Nikolson scheme')
plt.plot(st, c, label='BSM')
plt.xlabel("Value asset")
plt.ylabel("Value call option")
plt.grid()
plt.legend()
plt.show()
```
